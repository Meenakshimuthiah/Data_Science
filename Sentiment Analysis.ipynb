```python
import nltk
#Data imports
import numpy as np
import pandas as pd

#json
import json
```

```python
data = pd.read_json('C:/Users/meena/Documents/Information_Systems/test.json',orient='columns',lines=True)
data
```

We will analyzeif the yelp text reflects the stars given to it. We will be using only reviews with rating 1 and 5 to better check the results.

```python
yelp_data = data.drop(['business_id','cool','date','funny','review_id','type','useful','user_id'],1)
```

```python
yelp = yelp_data[(yelp_data['stars'] == 1) | (yelp_data['stars'] == 5)]
yelp
```

```python
X = yelp['text']
```

```python
Y = yelp['stars']
```

```python
from nltk.corpus import stopwords
import string
def text_process(text):
    nosplChar = [char for char in text if char not in string.punctuation]
    nosplChar = ''.join(nosplChar)
    
    return [word for word in nosplChar.split() if word.lower() not in stopwords.words('english')]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidvectorizer = TfidfVectorizer(ngram_range=(1,2))
vectors = tfidvectorizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(vectors, Y, test_size=0.33, random_state=42)
```

We will use support vector machines for our analysis

```python
from sklearn.svm import LinearSVC
svc = LinearSVC()
svc.fit(X_train, y_train)
```

```python
pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```
